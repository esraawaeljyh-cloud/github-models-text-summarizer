```python
from getpass import getpass
GITHUB_TOKEN = getpass("Paste your GitHub PAT (input is hidden): ")
print("Token stored in memory (not printed).")
```

```python
from openai import OpenAI
# Create client using GitHub Models endpoint
client = OpenAI(
base_url="https://models.github.ai/inference/v1",
api_key=GITHUB_TOKEN
)
print("Client created successfully.")
```

```python
response = client.chat.completions.create(
    model="microsoft/Phi-4-multimodal-instruct",
    messages=[
        {"role": "user", "content": "Hello from Google Colab! Write one fun fact about AI."}
    ]
)

print("Model response:\n")
print(response.choices[0].message.content)
```

```python
paragraph = """
Machine learning is a branch of artificial intelligence that focuses on building systems
that learn from data. Instead of being explicitly programmed, these systems identify
patterns and make decisions with minimal human intervention. Machine learning is widely
used in applications such as recommendation systems, fraud detection, and image recognition.
"""

prompt = f"""
Summarize the following paragraph into exactly 3 bullet points:

{paragraph}
"""

response = client.chat.completions.create(
    model="microsoft/Phi-4-multimodal-instruct",
    messages=[
        {"role": "user", "content": prompt}
    ],
    temperature=0.5,
    max_tokens=200
)

print("Original Paragraph:\n")
print(paragraph)

print("\nSummary (3 Bullet Points):\n")
print(response.choices[0].message.content)
```
